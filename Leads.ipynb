```python
# Importing necessary libraries for roc curve

from sklearn.metrics import roc_curve       
from sklearn.metrics import roc_auc_score


#Creating function for ROC curve
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
#Plotting  ROC curve for evluation of our model
draw_roc(y_pred_final_df.Converted,y_pred_final_df.Converted_prob)
```

- The roc curve is towards the left meaning our has good accuracy

- Area under the curve is 0.88 

```python
numbers = [float(x)/10 for x in range(10)]

for i in numbers:
    y_pred_final_df[i] = y_pred_final_df['Converted_prob'].map(lambda x:1 if x>i else 0 )
y_pred_final_df.head() 
```

```python
#Calculating accuracy, Sensitivity and specificty for evaluating optimal cut-off point

from sklearn.metrics import confusion_matrix

#dataframe to store the cut-offs
cutoff = pd.DataFrame(columns= ['Probability','Accuracy','Sensitvity','Specificity'])

for i in numbers:
    cm_matrix = confusion_matrix(y_pred_final_df['Converted'],y_pred_final_df[i])
    total=sum(sum(cm_matrix))                                                          # Taking the sum of the matrix
    accuracy=(cm_matrix[0,0]+cm_matrix[1,1])/total                                     # Storing Accuracy Data 
    sensitivity=cm_matrix[1,1]/(cm_matrix[1,0]+cm_matrix[1,1])                         # Storing Sensitivity Data
    specificity=cm_matrix[0,0]/(cm_matrix[0,0]+cm_matrix[0,1])                         # Storing Specificity Data
    cutoff.loc[i]=[i, accuracy, sensitivity, specificity]                          # Inserting all the data into the dataframe created earlier
print(cutoff)   # Printing the data
```

We willl choose the cutoff of 0.3 as for this value Accuracy  ,Sensitvity  ,Specificity are nearly equal.

```python
#Plotting a graph
cutoff.plot.line(x='Probability',y=['Accuracy','Sensitvity','Specificity'])
plt.show()
```

From the above graph we can see the optimal cutoff value is 0.3

```python
#calculating y-train keeping threshold 0.4
y_pred_final_df['Predicted']=y_pred_final_df['Converted_prob'].map(lambda x:1 if x >0.4 else 0 )  # Predicted value 

y_pred_final_df.head()
```
